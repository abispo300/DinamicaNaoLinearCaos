```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mapa_logistico(r,x,n):
    # A função retorna uma lista até n-ésima geração
    f_x = [x]
    for i in range(n):
        x = r*x*(1-x)
        f_x.append(x)
    return f_x
```

```python
def pontos_fixos(f_x):
    # A função retorna os pontos fixos a partir de uma lista
    # Cria um vetor com 4 digitos decimais a partir do mapa logistico
    f_xV = np.round(np.array(f_x),4)
    
    # Encontra os valores unicos e conta as recorrências
    valores_unicos, reco = np.unique(f_xV, return_counts=True)
    
    # Filtra os valores repetidos
    repetidos = valores_unicos[reco > 1]
    
    return repetidos
```

```python
a = mapa_logistico(3.9,0.1,100)
pontos_fixos(a)
```

```python
def grafico(r, x0, n, loc_):
    y = mapa_logistico(r, x0, n)
    ponto = pontos_fixos(y)
    
    if len(ponto) > 1:
        legenda = [f'Pontos fixos: {ponto}']
    elif len(ponto)==1:
        legenda = [f'Ponto fixo: {ponto[0]}']
    else:
        legenda = [f'Ponto fixo: {ponto.any()}']
    
    plt.figure(figsize=(10, 4))
    plt.plot(y, marker='o', color='b')
    plt.title(f'Mapa Logístico com taxa r = {r} e população inicial x0 = {x0}')
    plt.ylabel('População Reduzida')
    plt.xlabel('Tempo Discreto')
    plt.legend(handles=[plt.Line2D([], [], linestyle='None', label=legenda[0])], loc=loc_)
    return plt.show()
```

```python
grafico(0.5,0.1,100,'upper right')
```

```python
grafico(1.5,0.1,100,'lower right')
```
