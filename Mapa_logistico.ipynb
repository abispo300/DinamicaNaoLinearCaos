```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def mapa_logistico(r,x,n):
    f_x = [x]
    for i in range(n):
        x = r*x*(1-x)
        f_x.append(x)
    return f_x
```

```python
def pontos_fixos(f_x):
    # Cria um vetor com 4 digitos decimais a partir do mapa logistico
    f_xV = np.round(np.array(f_x),4)
    
    # Encontra os valores unicos e conta as recorrÃªncias
    valores_unicos, reco = np.unique(f_xV, return_counts=True)
    
    # Filtra os valores repetidos
    repetidos = valores_unicos[reco > 1]
    
    return repetidos
```

```python
r = 0.5
x0 = 0.1
n = 100
y = mapa_logistico(r,x0,n)
pontos = pontos_fixos(y)

plt.figure(figsize = (10,4))
plt.plot(y, marker = 'o', color = 'b')
plt.title(f'Mapa LogÃ­stico com taxa r = {r} e populaÃ§Ã£o inicial x0 = {x0}')
plt.ylabel('PopulaÃ§Ã£o Reduzida')
plt.xlabel('Tempo Discreto')

legenda = [f'Ponto Fixo: {pontos}']
plt.legend(handles=[plt.Line2D([], [], linestyle='None', label=legenda[0])], loc='upper right')

plt.show()
```
