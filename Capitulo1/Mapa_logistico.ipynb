```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme('paper', style='ticks', palette= 'bright' ) 
```

```python
def mapa_logistico(r,x,n):
    # A função retorna uma lista até n-ésima geração
    f_x = [x]
    for i in range(n):
        x = r*x*(1-x)
        f_x.append(x)
    return f_x
```

```python
def pontos_fixos(f_x,d_):
    # A função retorna os pontos fixos a partir de uma lista
    
    # Cria um vetor com d_ casas decimais a partir da lista f_x
    f_xV = np.round(np.array(f_x),d_)
    
    # Encontra os valores unicos e conta as recorrências
    valores_unicos, reco = np.unique(f_xV, return_counts=True)
    
    # Filtra os valores repetidos
    repetidos = valores_unicos[reco > 1]
    
    return repetidos
```

```python
def grafico(r, x0, n, loc_,d_):
    # Plota o mapa logístico com uma serie temporal discreta.
    # r: taxa do mapa; x0: população inical reduzida; n: número de iterações; loc_:posição da legenda; d_: casas decimais para o calculo dos pontos fixos
    
    y = mapa_logistico(r, x0, n)
    ponto = pontos_fixos(y,d_)
    
    if len(ponto) > 1:
        legenda = [f'Pontos fixos: {ponto}']
    elif len(ponto)==1:
        legenda = [f'Ponto fixo: {ponto[0]}']
    else:
        legenda = [f'Não há ponto fixo']
    
    plt.figure(figsize=(12, 2))
    plt.plot(y, marker='o', color='k')
    plt.title(f"Mapa logístico com taxa $r$ = {r} e população inicial $x_0$ = {x0}", fontstyle='italic')
    plt.ylabel('População Reduzida')
    plt.xlabel('Tempo Discreto')
    plt.xlim(left=0)
    sns.despine(trim=True)
    plt.legend(handles=[plt.Line2D([], [], linestyle='None', label=legenda[0])], loc=loc_)
    return plt.show()
```

```python
grafico(0.5,0.1,100,'upper right',4)
```
