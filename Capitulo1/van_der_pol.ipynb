```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import seaborn as sns
sns.set_theme('paper', style='ticks', palette= 'bright' ) 
```

```python
def van_der_pol(t, z, e):
    x, y = z
    dxdt = y
    dydt = -x + e * (1 - x**2) * y
    return [dxdt, dydt]
```

```python
# tempo da simulação
t_span = (0, 650)
t_eval = np.linspace(*t_span, 100000)
```

```python
initial_conditions = [(1, 0),(0, 1),(0.5, 0.5),(-0.5, -0.5),(0.1, -0.1)]
e = 1
solutions = []
for ci in initial_conditions:
    sol = solve_ivp(van_der_pol, t_span, ci, args=(e,), t_eval=t_eval, rtol=1e-9)
    solutions.append(sol)
```

```python
plt.figure(figsize=(8, 6))
for sol in solutions:
    plt.plot(sol.y[0], sol.y[1], label=f'CI: {sol.y[0][0]:.2f}, {sol.y[1][0]:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Ciclo-Limite Atrativo no Plano de Fase - Van der Pol', fontstyle = 'italic')
plt.legend()
sns.despine()

# Definindo a grade de pontos para o campo de vetores
x = np.linspace(-3, 3, 25)
y = np.linspace(-3, 3, 25)
X, Y = np.meshgrid(x, y)
U, V = van_der_pol(0, [X, Y], e)
plt.quiver(X, Y, U, V, color='k', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(6, 3))

# x(t) vs. tempo
plt.subplot(1, 2, 1)
plt.plot(sol.t, sol.y[0], color = 'k')
plt.xlabel('t')
plt.xlim(600, 650)
plt.ylabel('x(t)')
plt.ylim(-3, 3)
plt.legend(title='$\epsilon$ = 10')

# y(t) vs. tempo
plt.subplot(1, 2, 2)
plt.plot(sol.t, sol.y[1],  color = 'k')
plt.xlabel('t')
plt.xlim(600, 650)
plt.ylabel('y(t)')
plt.ylim(-3, 3)
plt.legend(title='$\epsilon$ = 10', loc = 'upper left')

plt.tight_layout()
plt.show()
```
