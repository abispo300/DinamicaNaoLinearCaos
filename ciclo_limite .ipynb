```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definir as equaes diferenciais
def system(t, state):
    x, y = state
    dxdt = y
    dydt = -x + (1 - x**2) * y
    return [dxdt, dydt]

# Condies iniciais
initial_conditions = [
    (1.0, 1.0),
    (-1.0, 1.0),
    (0.5, -1.0),
    (-0.5, -1.0)
]

# Tempo de integrao
t_span = (0, 100)  # de 0 a 100 unidades de tempo

# Resolver o sistema para cada condio inicial
plt.figure(figsize=(10, 8))
for ic in initial_conditions:
    sol = solve_ivp(system, t_span, ic, t_eval=np.linspace(t_span[0], t_span[1], 1000))

    # Plotar a trajetria no plano de fase (x, y)
    plt.plot(sol.y[0], sol.y[1], label=f'x0={ic[0]}, y0={ic[1]}')

# Decorar o grfico
plt.title('Plano de Fase (x, y) para $\epsilon = 1$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```
